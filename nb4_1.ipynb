```python
# 1) 문자 채우기, 정렬
name = "Python"

print(f"{name:*<15}")  # 왼쪽 정렬, 빈칸은 *로 채움
print(f"{name:*^15}")  # 가운데 정렬, 빈칸은 *로 채움
print(f"{name:*>15}")  # 오른쪽 정렬, 빈칸은 *로 채움
# 변수 사용 가능
w = 15
print(f"{name:*>{w}}")  # 오른쪽 정렬, 빈칸은 *로 채움
```

```python
name = "Park"
age = 30

# f-string 기본 사용
print(f"이름: {name}, 나이: {age}")  # 변수 값을 문자열 안에 직접 삽입

# 소수점 자리수 지정
pi = 3.141592
print(f"{pi:.2f}")  # 소수점 둘째 자리까지 출력 → 3.14

# 자리수(폭) 지정
num = 7
print(f"{num:5}")   # 전체 5칸 확보, 오른쪽 정렬 → '    7'
print(f"{num:05}")  # 빈칸을 0으로 채움 → 00007

# 왼쪽 정렬 (<), 오른쪽 정렬 (>), 가운데 정렬 (^)
print(f"{'Python':<10}")  # 왼쪽 정렬
print(f"{'Python':>10}")  # 오른쪽 정렬
print(f"{'Python':_^10}")  # 가운데 정렬
```

```python
# f-string 문법이 나오기 전에
# 기본 사용
print("이름: {}, 나이: {}".format("Park", 30))  # {} 순서대로 값 들어감

# 인덱스 사용
print("이름: {0}, 나이: {1}".format("Park", 30))  # 번호로 위치 지정

# 이름 지정
print("이름: {name}, 나이: {age}".format(name="Park", age=30))  
# 변수 이름으로 직접 지정 가능
```

```python
name = "Park"
age = 30

print("이름: %s, 나이: %d" % (name, age))  
# %s = 문자열, %d = 정수

pi = 3.141592
print("%.2f" % pi)  # 소수점 둘째 자리까지 출력 → 3.14
```

```python
# 9) 백분율 표시
rate = 0.8746
print(f"{rate:.0%}")  # 100을 곱하고 반올림
print(f"{rate:.1%}")  # 100을 곱하고 소수점 첫째 자리까지 반올림 % 표시
print(f"{rate:.2%}")

# 87%
# 87.5%
# 87.46%
```

```python
# 10) 2진수, 8진수, 16진수 
num = 255
print(f"{num:b}")  # 11111111 2진수
print(f"{num:o}")  # 377      8진수
print(f"{num:x}")  # ff       16진수 소문자
print(f"{num:X}")  # FF       16진수 대문자
```

```python
# 11) 날짜 포맷팅
# 파이썬 언어에 포함된 객체/함수라도 내장객체/함수와 모듈안에 포함된 객체/함수로 구분된다.
# 모듈안에 포함된 것들은 import가 필요하다. 뒤에서 모듈수업때 설명
from datetime import datetime # datetime 모듈(module) 안에 있는 datetime 클래스(객체)를 가져온다.
today = datetime.now()
print(f"{today:%Y-%m-%d}")          # 연-월-일
print(f"{today:%H:%M:%S}")          # 시:분:초
print(f"{today:%Y/%m/%d %H:%M}")    # 날짜 + 시간
# 2026-06-22
# 19:48:14
# 2026/06/22 19:48
```
