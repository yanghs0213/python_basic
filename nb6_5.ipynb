```python
for student in students:

    gender_code = student["주민번호"].split("-")[1][0]

    if gender_code in ["1", "3"]:
        student["성별"] = "남"
    else:
        student["성별"] = "여"

students
```

```python
from datetime import datetime

current_year = datetime.now().year

for student in students:
    rrn = student["주민번호"]
    yy = int(rrn[:2])
    code = rrn.split("-")[1][0]

    if code in ["1", "2"]:
        birth_year = 1900 + yy
    else:
        birth_year = 2000 + yy

    student["나이"] = current_year - birth_year

students
```

```python
scores = []

for student in students:
    scores.append(student["성적"])

print("개수:", len(scores))
print("합계:", sum(scores))
print("평균:", sum(scores) / len(scores))
print("최대:", max(scores))
print("최소:", min(scores))
```

```python
age_group = {}

for student in students:

    decade = (student["나이"] // 10) * 10

    if decade not in age_group:
        age_group[decade] = []

    age_group[decade].append(student["성적"])

print(age_group)
```

```python
for decade, scores in age_group.items():
    avg = sum(scores) / len(scores)
    print(f"{decade}대 평균: {avg:.2f}")
```

```python
# 이상치 탐지

for order in orders:
    price = order["구매금액"]

    if price is None:
        continue

    if price < 0 or price > 1000000:
        print(f"{order}, 구매금액: {price:,}")
```

```python
# 이상치 처리

for order in orders:
    price = order["구매금액"]

    if price is None:
        continue

    if price > 1000000:
        order["구매금액"] = 1000000
        print(f"{order['구매금액']:,}")

    elif price < 0:
        order["구매금액"] = 0
        print(f"{order['구매금액']:,}")
```

```python
# 결측치 처리

prices = []

for order in orders:
    if order["구매금액"] is not None:
        prices.append(order["구매금액"])

avg = round(sum(prices) / len(prices))

for order in orders:
    if order["구매금액"] is None:
        order["구매금액"] = avg
   
for order in orders:
    temp = order.copy()
    temp["구매금액"] = f"{temp['구매금액']:,}"
    print(temp)
```

```python
# 회원 가입 연수 계산 (특성 추출)

from datetime import datetime

current_date = datetime.now()

for order in orders:
    join_date = datetime.strptime(order["가입일"], "%Y-%m-%d")
    order["가입연수"] = current_date.year - join_date.year

orders
```

```python
# VIP 여부 생성 (특성 추출)

for order in orders:
    price = order["구매금액"]

    if price >= 200000:
        order["회원등급"] = "VIP"
    else:
        order["회원등급"] = "일반"

orders
```

```python
# 기술통계

prices = [order["구매금액"] for order in orders]

print("개수:", len(prices))
print("합계:", f"{sum(prices):,}")
print("평균:", f"{round(sum(prices) / len(prices)):,}")
print("최대값:", f"{max(prices):,}")
print("최소값:", f"{min(prices):,}")

seoul_sum = 0
seoul_count = 0

busan_sum = 0
busan_count = 0

daegu_sum = 0
daegu_count = 0

gwangju_sum = 0
gwangju_count = 0

daejeon_sum = 0
daejeon_count = 0

for order in orders:

    if order["지역"] == "서울":
        seoul_sum += order["구매금액"]
        seoul_count += 1

    elif order["지역"] == "부산":
        busan_sum += order["구매금액"]
        busan_count += 1

    elif order["지역"] == "대구":
        daegu_sum += order["구매금액"]
        daegu_count += 1

    elif order["지역"] == "광주":
        gwangju_sum += order["구매금액"]
        gwangju_count += 1

    elif order["지역"] == "대전":
        daejeon_sum += order["구매금액"]
        daejeon_count += 1

print("서울 평균:", f"{round(seoul_sum / seoul_count):,}")
print("부산 평균:", f"{round(busan_sum / busan_count):,}")
print("대구 평균:", f"{round(daegu_sum / daegu_count):,}")
print("광주 평균:", f"{round(gwangju_sum / gwangju_count):,}")
print("대전 평균:", f"{round(daejeon_sum / daejeon_count):,}")
```

```python
# 회원등급별 평균 구매금액

vip_sum = 0
vip_count = 0

normal_sum = 0
normal_count = 0

for order in orders:

    if order["회원등급"] == "VIP":
        vip_sum += order["구매금액"]
        vip_count += 1

    else:
        normal_sum += order["구매금액"]
        normal_count += 1

print("VIP 평균:", f"{round(vip_sum / vip_count):,}")
print("일반 평균:", f"{round(normal_sum / normal_count):,}")
```

```python
# 가입연수별 그룹 분석

g1_sum = g1_count = 0
g2_sum = g2_count = 0
g3_sum = g3_count = 0

for order in orders:

    years = order["가입연수"]
    price = order["구매금액"]

    if years <= 2:
        g1_sum += price
        g1_count += 1

    elif years <= 5:
        g2_sum += price
        g2_count += 1

    else:
        g3_sum += price
        g3_count += 1

print("0~2년 평균:", f"{round(g1_sum / g1_count):,}")
print("3~5년 평균:", f"{round(g2_sum / g2_count):,}")
print("6년 이상 평균:", f"{round(g3_sum / g3_count):,}")
```

```python
# 가장 구매금액이 높은 고객 TOP 3

customers = []

for order in orders:
    customers.append(order)

sorted_customers = sorted(customers, key=lambda x: x["구매금액"], reverse=True)

count = 0

for customer in sorted_customers:

    print("고객명:", customer["고객명"],
          "구매금액:", f"{customer['구매금액']:,}")
   
    count += 1

    if count == 3:
        break
```
