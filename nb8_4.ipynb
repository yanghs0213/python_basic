```python
with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

    def is_outlier(x):
        if x in ["", "결측", "NaN"]:
            return True
        x = int(x)
        return x < 0 or x > 100

    count = 0

    for s in data:
        # 나이
        if is_outlier(s["나이"]):
            count = count + 1
        # 국어   
        if is_outlier(s["국어"]):
            count = count + 1
        # 영어
        if is_outlier(s["영어"]):
            count = count + 1

print(count)
```

```python
with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

    count = 0

    for s in data:
        if s["국어"] in ["결측", "NaN", ""]:
            count = count + 1
        if s["영어"] in ["결측", "NaN", ""]:
            count = count + 1

print(count)
```

```python
with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

    kor_total = 0
    kor_count = 0
    eng_total = 0
    eng_count = 0

    for s in data:
        if s["국어"] not in ["결측", "NaN", ""]:
            kor_total = kor_total + int(s["국어"])
            kor_count = kor_count + 1
        if s["영어"] not in ["결측", "NaN", ""]:
            eng_total = eng_total + int(s["영어"])
            eng_count = eng_count + 1

    kor_avg = round(kor_total / kor_count, 1)
    eng_avg = round(eng_total / eng_count, 1)

    print(kor_avg)
    print(eng_avg)

    for s in data:
        if s["국어"] in ["결측", "NaN", ""]:
            s["국어"] = kor_avg
        if s["영어"] in ["결측", "NaN", ""]:
            s["영어"] = eng_avg

    for s in data:
        print(s)
```
