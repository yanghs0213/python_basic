```python
import csv

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

def is_outlier(x):
    if x in ["", "결측", "NaN"]:
        return True
    x = int(x)
    return x < 0 or x > 100

age_sum = age_cnt = 0
kor_sum = kor_cnt = 0
eng_sum = eng_cnt = 0

for s in data:
    if not is_outlier(s["나이"]):
        age_sum += int(s["나이"])
        age_cnt += 1
    if not is_outlier(s["국어"]):
        kor_sum += int(s["국어"])
        kor_cnt += 1

    if not is_outlier(s["영어"]):
        eng_sum += int(s["영어"])
        eng_cnt += 1

age_mean = round(age_sum / age_cnt)
kor_mean = round(kor_sum / kor_cnt)
eng_mean = round(eng_sum / eng_cnt)

print(age_mean, kor_mean, eng_mean)

for s in data:
    if is_outlier(s["나이"]):
        s["나이"] = round(age_mean)

    if is_outlier(s["국어"]):
        s["국어"] = round(kor_mean)

    if is_outlier(s["영어"]):
        s["영어"] = round(eng_mean)

for s in data:
    print(s)
```

```python
import csv

def safe_int(x):
    if x in ["", "결측", "NaN"]:
        return None
    return int(x)

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

total_sum = 0

for s in data:
    kor = safe_int(s["국어"])
    eng = safe_int(s["영어"])

    if kor is None:
        kor = 0
    if eng is None:
        eng = 0

    # 학생별 총점
    total = kor + eng
    s["총점"] = total

    # 학생별 평균
    s["평균"] = total / 2

    # 전체 학생 총점 합계
    total_sum += total

# 전체 학생 총점 평균
total_mean = total_sum / len(data)

for s in data:
    print(f'{s["이름"]} : 총점={s["총점"]}, 평균={s["평균"]}')

print("전체 학생 총점:", total_sum)
print("전체 학생 총점 평균:", total_mean)
```

```python
import csv

def safe_int(x):
    if x in ["", "결측", "NaN"]:
        return None
    return int(x)

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

for s in data:
    kor = safe_int(s["국어"])
    eng = safe_int(s["영어"])

    if kor is None:
        kor = 0
    if eng is None:
        eng = 0

    total = kor + eng
    s["총점"] = total
    s["평균"] = total / 2

for s in data:
    if s["평균"] >= 60:
        s["합격여부"] = "합격"
    else:
        s["합격여부"] = "불합격"

for current_student in data:
    rank = 1
    for other_student in data:
        if other_student["총점"] > current_student["총점"]:
            rank += 1
    current_student["순위"] = rank

for s in data:
    print(f'{s["이름"]} : 총점={s["총점"]}, 평균={s["평균"]}, {s["합격여부"]}, 순위={s["순위"]}')
```
