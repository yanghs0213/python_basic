```python
# 빈 딕셔너리
d1 = {}
d2 = dict()

print(d1)
print(d2)
```

```python
# {} 사용
student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}

print(student)
print(type(student))
```

```python
# dict() 사용
person = dict(name="김철수", age=30, city="서울")

print(person)
```

```python
# [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]

fruit = dict(data)

print(fruit)
```

```python
d = {"name": "홍길동"}
print(d)

d = {100: "학생"}
print(d)

d = {
    (1, 2): "좌표"
}
print(d)

d = {
    [1, 2]: "리스트"
} # error
print(d)
```

```python
student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제 del문 : 반환값 없다
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```

```python
student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 90})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```
