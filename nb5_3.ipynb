```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50]

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::2])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[1::2])

nums = [10, 20, 30, 40, 50]
print(nums[::-1])

nums = [10, 20, 30, 40, 50, 60]
print(nums[-3:])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-4::2])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::-2])

nums = [10, 20, 30, 40, 50, 60, 70]
print(nums[5:1:-1]) # 5위치부터 1전까지 -1칸식  [60, 50, 40, 30]

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-1:-6:-1])

nums = [10, 20, 30, 40, 50]
nums[1:4] = [7, 8] # 1~3구간 [20, 30, 40]에 [7, 8]을 대입 
print(nums) # [10, 7, 8, 50]

a = [10, 20, 30, 40, 50, 60]
b = a[:]
c = a[::-1]
d = a[1::2]
e = a[:2]

print(b)
print(c)
print(d)
print(e)

print(nums)
```
