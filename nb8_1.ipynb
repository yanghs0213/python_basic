```python
greet = "Hello"
while greet: # 문자열의 경우 공백이면 False, 아니면 True
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10
# 다중 할당
# 내부적으로 오른쪽의 튜플 (0, 10)을 왼쪽의 변수들에 언패킹해서 각각 저장

while a < b:
    print(a)
    a = a + 1 # a += 1
```

```python
try:
    num = int(input("숫자 입력 : "))
    result = 10 / num
    print(result)

    print("자원해지")

except ValueError:
    print("숫자를 입력하세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

finally:
    print("프로그램 종료")

print("종료")
```
