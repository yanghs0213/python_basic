```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

even = []

for i in list_nums:
    if i%2 == 0:
        even.append(i)

print(even)

even2 = [i for i in list_nums if i%2 == 0]

print(even2)

print([i*10 for i in list_nums])
print([i*10 for i in list_nums if i<=5])

names = ["홍길동", "이순신", "이황"]

print([n+"님" for n in names])

print([n+"님" for n in names if n.startswith("이")])

names2 = ["홍길동", "이순신", "김유신", "박혁거세"] # n[:1], n[1:]
print({n[:1]:n[1:]for n in names2})
```
